# 处理数据

```python
import pandas as pd
import numpy as np
```

## 1、kkb_commodity 商品表

```python
kkb_commodity = pd.read_csv('./kkb_commodity.csv',encoding='utf-8')

# 添加列索引
kkb_commodity.columns = ['sp_type','sp_id', 'sp_name','sp_sku','sp_spu','price','purch_price',
                         'sales_status','online_date','offline_date','online_operator','offline_operator','event_day']
                         
kkb_commodity.tail()
```

```python
kkb_commodity.info()
```

```python
kkb_commodity.describe()
```

```python
# 转换日期格式
kkb_commodity['event_day'] = pd.to_datetime(kkb_commodity.event_day,format='%Y-%m-%d')

# 选取2020-01-01至2020-01-07 之间的数据
df = kkb_commodity.query('event_day >= "2020-01-01" and event_day<= "2020-01-07"')

df.to_csv('./kkb_commodity_2020-01-1to7.csv',encoding='utf-8',index=False)

df.tail()
```
